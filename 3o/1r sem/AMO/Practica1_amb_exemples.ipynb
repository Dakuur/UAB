```python
import scipy
import numpy as np
import sympy as sp
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

# Exemple 1:

```python
f = sp.Function('f')
x = sp.Symbol('x')

f = x**2 + 4*x + 4

# Trobar el mínim global de la funció f(x):
f_prime = f.diff(x)

minim = sp.solve(f_prime, x)
print(f"Mínims: {minim}")
if len(minim) == 1:
    print(f"El mínim global de la funció f(x) és a x={minim[0]} i equival a {f.subs(x, minim[0])}")
```

```python
def graficar(funcion_sympy, variable):
    # Derivar la función
    primera_derivada = sp.diff(funcion_sympy, variable)
    segunda_derivada = sp.diff(primera_derivada, variable)
    
    # Encontrar los puntos críticos
    puntos_criticos = sp.solve(primera_derivada, variable)
    
    # Crear un rango de valores para graficar la función
    x_vals = np.linspace(-10, 10, 400)
    y_vals = [funcion_sympy.subs(variable, val) for val in x_vals]
    
    # Graficar la función
    plt.plot(x_vals, y_vals, label='Función')
    
    # Evaluar los puntos críticos y determinar si son mínimos o máximos
    for punto in puntos_criticos:
        valor_segunda_derivada = segunda_derivada.subs(variable, punto)
        if valor_segunda_derivada > 0:
            # Mínimo
            plt.plot(punto, funcion_sympy.subs(variable, punto), 'go', label='Mínimo')
        elif valor_segunda_derivada < 0:
            # Máximo
            plt.plot(punto, funcion_sympy.subs(variable, punto), 'ro', label='Máximo')
    
    # Mostrar la gráfica
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.title('Gráfico de la función con puntos críticos')
    plt.grid(True)
    plt.show()
```

```python
def comprovar(f):
    result = minimize(f, x0=0, constraints=None)
    print("Mínim global en x =", result.x)
    print("Valor de la funció en el mínim =", result.fun)
```

# Problema 1

```python
from sympy import symbols, diff, solve

# Definir la variable y la función
x = symbols('x')
f =(x-2)**2

# Derivadas de primer y segundo orden
df = diff(f, x)
d2f = diff(df, x)

# Encontrar puntos críticos
critical_points = solve(df, x)

# Analizar puntos críticos
for point in critical_points:
    hessian_value = d2f.subs(x, point)  # Usamos la segunda derivada para una función de una variable
    print(f"\nPunto crítico: {point}")
    print(f"Valor de la segunda derivada: {hessian_value}")
    if hessian_value > 0:
        print("Es un mínimo local.")
    elif hessian_value < 0:
        print("Es un máximo local.")
    else:
        print("Es un punto de silla.")
        
graficar(f, x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
# Definir els punts
punts = np.array([[0, 0], [1, 0], [0, 1], [0.5, 0.5]])
# Crear el conjunt convex
hull = ConvexHull(punts)
# Dibuixar els punts i el conjunt convex
plt.plot(punts[:, 0], punts[:, 1], 'o')
for simplex in hull.simplices:
    plt.plot(punts[simplex, 0], punts[simplex, 1], 'k-')
    
plt.fill(punts[hull.vertices, 0], punts[hull.vertices, 1], 'lightblue', alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Conjunt Convex')
plt.show()
```

```python
from scipy.optimize import minimize

# Definir la funció objectiu
def func_objectiu(x):
    return x[0]**2 + x[1]**2

# Definir la restricció
def restriccio(x):
    return 1 - (x[0] + x[1])

# Definir els límits i les restriccions
restriccions = {'type': 'ineq', 'fun': restriccio}
x0 = [0.5, 0.5]  # Punt inicial

# Resoldre el problema d’optimització
sol = minimize(func_objectiu, x0, constraints=restriccions)

# Mostrar el resultat
print('Solució òptima:', sol.x)
print('Valor de la funció objectiu:', sol.fun)
```
