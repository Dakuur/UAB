```python
from dataloader_module import load_data
import matplotlib.pyplot as plt
import seaborn as sns

datafile = "../data/healthcare-dataset-stroke-data.csv"
X, y = load_data(datafile)

X.head()
```

```python
df = X.copy()
df["stroke"] = y

selected_features = ["age", "avg_glucose_level", "bmi"]
num_features = len(selected_features)
num_cols = 3
num_rows = (num_features + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 7))
axes = axes.flatten()

for i, c in enumerate(selected_features):
    sns.violinplot(x='stroke', y=c, data=df, ax=axes[i], hue='stroke', palette="Set2", split=True, legend=False)
    sns.stripplot(x='stroke', y=c, data=df, ax=axes[i], color='black', alpha=0.3, size=2)
    axes[i].set_title(c.capitalize(), fontsize=16, fontweight='bold')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
stroke_count = df[df['stroke'] == 1].shape[0]
total_count = df.shape[0]
stroke_percentage = (stroke_count / total_count) * 100

print(f"Número de files on stroke=1: {stroke_count}")
print(f"Percentatge de files on stroke=1: {stroke_percentage}%")
print(f"Número total de files: {total_count}")
```
