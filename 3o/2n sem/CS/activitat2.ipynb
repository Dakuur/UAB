### 1. Make a program that for a given number n tells wether it is prime or not looking for factors in [1 : n].

```python
from time import time

def is_prime1(n):
    if type(n) != int:
        raise TypeError("Input must be an integer")
    if n < 2:
        raise ValueError("Input must be greater than 1")
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for i in range(2, 11):
    print(i, is_prime1(i))

start = time()
for i in range(2, 10000):
    prime = is_prime1(i)
print("\nTime:", time()-start)
```

### 2. Make a program that for a given number n tells wether it is prime or not looking for factors in [1 : n/2].

```python
def is_prime2(n):
    if type(n) != int:
        raise TypeError("Input must be an integer")
    if n < 2:
        raise ValueError("Input must be greater than 1")
    for i in range(2, int(n/2)+1):
        if n % i == 0:
            return False
    return True

for i in range(2, 11):
    print(i, is_prime2(i))

start = time()
for i in range(2, 10000):
    prime = is_prime1(i)
print("\nTime:", time()-start)
```

### 3. Make a program that for a given number n tells wether it is prime or not using Fermat’s theorem.

```python
import random

def is_prime_fermat(n, k=15):
    if type(n) != int:
        raise TypeError("Input must be an integer")
    if n < 4:
        raise ValueError("Input must be greater than 4")
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for _ in range(k):
        a = random.randint(2, n - 2)
        if pow(a, n - 1, n) != 1:
            return False
    return True

for i in range(4, 15):
    print(i, is_prime_fermat(i))

start = time()
for i in range(10, 10000):
    prime = is_prime_fermat(i, 10)
print("\nTime:", time()-start)
```

### 4. Run your programs for one minute looking for primes.

```python
def run_for_one_minute(func, start=2):
    max_prime = None
    n = start
    start_time = time()
    while time() - start_time < 60:
        if func(n):
            max_prime = n
        n += 1
    return max_prime

max_prime1 = run_for_one_minute(is_prime1)
print("Maximum prime found by is_prime1 in 1 minute:", max_prime1)

max_prime2 = run_for_one_minute(is_prime2)
print("Maximum prime found by is_prime2 in 1 minute:", max_prime2)

max_prime_fermat = run_for_one_minute(is_prime_fermat, start=4)
print("Maximum prime found by is_prime_fermat in 1 minute:", max_prime_fermat)
```

### 5. Draw a graph (x-axis number, y-axis time) using both programs to calculate the time spent for every number.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of numbers
numbers = range(4, 10_000)

# Measure time for is_prime1
times_is_prime1 = []
for n in numbers:
    start = time()
    is_prime1(n)
    times_is_prime1.append(time() - start)

# Measure time for is_prime2
times_is_prime2 = []
for n in numbers:
    start = time()
    is_prime2(n)
    times_is_prime2.append(time() - start)

# Measure time for is_prime_fermat
times_is_prime_fermat = []
for n in numbers:
    start = time()
    is_prime_fermat(n)
    times_is_prime_fermat.append(time() - start)

# Plot the results
plt.plot(numbers, np.array(times_is_prime1)*1000, label='is_prime1', color='blue', marker='o')
plt.plot(numbers, np.array(times_is_prime2)*1000, label='is_prime2', color='green', marker='o')
plt.plot(numbers, np.array(times_is_prime_fermat)*1000, label='is_prime_fermat', color='red', marker='o')

# Add labels, legend, and title
plt.xlabel('Number')
plt.ylabel('Time (milliseconds)')
plt.title('Time Spent for Each Number')
plt.legend()

# Show the graph
plt.show()
```
