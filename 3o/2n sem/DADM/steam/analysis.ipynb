```python
import pandas as pd
import matplotlib.pyplot as plt

filename = "Valve_Player_Data.csv"

# Load the CSV file into a DataFrame
data = pd.read_csv(filename)

# Convert the 'Month_Year' column to datetime format
data['Month_Year'] = pd.to_datetime(data['Month_Year'], format='%B %Y')

# Sort the data by 'Month_Year'
data = data.sort_values('Month_Year')

# Select the top 5 games to include in the plot
top_games = ['Counter Strike: Global Offensive', 'Dota 2', "PAYDAY 2", "ARK: Survival Evolved", "Dead by Daylight"]

# Filter the data to include only the top 5 games
filtered_data = data[data['Game_Name'].isin(top_games)]

# Plot the data
plt.figure(figsize=(10, 6))
for game in top_games:
    game_data = filtered_data[filtered_data['Game_Name'] == game]
    plt.plot(game_data['Month_Year'], game_data['Avg_players'], label=game)

plt.xlabel('Month')
plt.ylabel('Number of Players')
plt.title('Steam Game Activity Over Time')
plt.legend()
plt.show()
```
