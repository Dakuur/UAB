```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

edges = [    ("Cat", "Sagrera", 13),    ("Cat", "Andreu", 14),    ("Cat", "Sants", 8),    ("Cat", "Provena", 3),    ("Sants", "Castellbisbal", 32),    ("Sants", "Provena", 9),    ("Provena", "UAB", 35),    ("Provena", "Sagrera", 17),    ("Castellbisbal", "UAB", 16),    ("Sagrera", "UAB", 30),    ("Sagrera", "Andreu", 2),    ("Andreu", "UAB", 21)]

G.add_weighted_edges_from(edges)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300) # set node_size to 300
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()

matrix = nx.adjacency_matrix(G, weight='weight')
print(matrix.todense())
```
