```python
total = 0

def f(n):
    global total

    if n == 0:
        return 1
    total += 1
    sum = 0
    for i in range(n):
        sum += f(n-1)
    return sum

n = 12
print(f"N: {n}")

#f(n)

#print(f"Recursions: {total}")

#print(f"{n} factorial: {factorial(n)}")

array1 = []
for i in range(0, n):
    total = 0
    f(i)
    array1.append(total)
```

```python
import matplotlib.pyplot as plt
from math import factorial

# Sample arrays

array2 = [factorial(i) for i in range(0, n)]

# Plotting
plt.plot(array1, label='number of recursions')
plt.plot(array2, label='theoretical n!')

# Adding labels and legend
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of two arrays (logarythmic scale)')
plt.legend()

plt.yscale('log')

# Show plot
plt.show()
```
