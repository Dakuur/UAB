## Exercici 1

En aquest exercici s'haurà d'implementar les tres següents funcions per tal de poder convertir la imatge RGB a HSI, fent servir els càlculs de la intensitat, la saturació i del matís (hue). Un cop implementades totes les funcions caldrà executar la cel·la de sota per la visualització de les imatges.

```python
#Calcular Intensitat
def calcular_intensitat(r, b, g):
    return 1/3*(r+b+g)
    
#Calcular Saturació
def calcular_saturacio(r, b, g):
    #Codi a implementar
    return 1/3*(r+b+g)

#Calculate Hue
def calcular_hue(r, b, g):
    #Codi a implementar
    return 1/3*(r+b+g)

```

```python
def rgb_a_hsi(img):
    with np.errstate(divide='ignore', invalid='ignore'):
        bgr = np.float32(img)/255
        b = bgr[:,:,0]
        g = bgr[:,:,1]
        r = bgr[:,:,2]

        hsi = cv2.merge((calcular_hue(r, g, b), calcular_saturacio(r, g, b), calcular_intensitat(r, g, b)))
        return hsi

img = cv2.imread('imatge.jpg', cv2.IMREAD_COLOR)
hsi = rgb_a_hsi(img)

rgb_hsi = cv2.cvtColor(hsi, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(rgb_hsi)
plt.title('HSI')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(hsi[:, :, 0], cmap='hsv')
plt.title('Hue')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(hsi[:, :, 1], cmap='gray')
plt.title('Saturació')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(hsi[:, :, 2], cmap='gray')
plt.title('Intensitat')
plt.axis('off')

plt.show()
```
